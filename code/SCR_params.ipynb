```python
import pandas as pd
import numpy as np
```

```python
subjects = ['E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M']
task_list = ['task1','task2','task3','task4']

df_amplitude = pd.DataFrame()
df_risetime = pd.DataFrame()
df_recoverytime = pd.DataFrame()


for task in task_list:
    amplitude = []
    risetime = []
    recoverytime = []
    
    for subject in subjects:
        df_scr = pd.read_csv("../data/" + subject + "/SCR/" + task + "_modified.csv")


        if len(df_scr)==0:
            amplitude.append(0)
            risetime.append(0)
            recoverytime.append(0)
        else:
            amplitude.append(np.mean(df_scr['Amplitude'].values))
            risetime.append(np.mean(df_scr['RiseTime'].values))
            recoverytime.append(np.mean(df_scr['HalfRecoveryTime'].values))
        
    df_amplitude[task] = amplitude
    df_risetime[task] = risetime
    df_recoverytime[task] = recoverytime 

df_amplitude.index = subjects
df_risetime.index = subjects
df_recoverytime.index = subjects
```
