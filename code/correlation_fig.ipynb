```python
import pandas as pd

task_list = ['task1', 'task2', 'task3','task4']
subject = 'F'

df_data = pd.DataFrame()
for task in task_list:
    df = pd.read_csv("../time_series/" + subject + "/" + task + ".csv")
    #df = df.iloc[len(df)-len(df)//30*30::30]
    df_data = pd.concat([df_data, df])

df_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cor = df_data.drop(['Unnamed: 0', 't', 'error', 'PE', 'NPE', 'PEm', 'NPEm', 'DMS', 'EDA'], axis=1).corr()
plt.figure(figsize=(20, 20))
sns.heatmap(cor, cmap= sns.color_palette('coolwarm', 10), annot=True,fmt='.2f', vmin = -1, vmax = 1)
```

```python
sns.pairplot(df_data.drop(['Unnamed: 0', 't', 'error', 'PE', 'NPE', 'PEm', 'NPEm', 'DMS', 'EDA'], axis=1))
```
